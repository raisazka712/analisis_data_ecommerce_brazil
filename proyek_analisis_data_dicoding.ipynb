### Pertanyaan 1 : Bagaimana tren penjualan di Olist Store selama periode 2016 hingga 2018?

#### Gabungkan Data
Untuk mendapatkan pandangan yang komprehensif, kita perlu menggabungkan beberapa tabel, terutama df_orders, df_order_items, dan df_order_payments.


```python
# Menggabungkan tabel orders dengan order items
df_orders_items = pd.merge(df_orders, df_order_items, on='order_id')
```

```python
# Menggabungkan df_orders_items dengan order_payments
df_orders_items_payments = pd.merge(df_orders_items, df_order_payments, on='order_id')
```

```python
# cek tabel nya
df_orders_items_payments
```

#### Tentukan Periode Waktu
Filter data untuk periode 2016 hingga 2018.

```python
# Filter data untuk periode 2016 hingga 2018.
df_orders_items_payments['order_purchase_timestamp'] = pd.to_datetime(df_orders_items_payments['order_purchase_timestamp'])
df_filtered = df_orders_items_payments[(df_orders_items_payments['order_purchase_timestamp'] >= '2016-01-01') & (df_orders_items_payments['order_purchase_timestamp'] <= '2018-12-31')]
```

```python
# cek datanya
df_filtered
```

####  Kalkulasi Total Penjualan
Hitung total penjualan per bulan untuk setiap tahun.

```python
# Menambahkan kolom bulan dan tahun
df_filtered['year_month'] = df_filtered['order_purchase_timestamp'].dt.to_period('M')

# Menghitung total penjualan
monthly_sales = df_filtered.groupby('year_month')['payment_value'].sum().reset_index()
monthly_sales['year_month'] = monthly_sales['year_month'].astype(str)
monthly_sales
```

```python
# urutkan data penjualannya
monthly_sales.sort_values(by="payment_value", ascending=False)
```

* findings :
  * Berdasarkan data diatas, dapat disimpulkan bahwa dataset penjualan dimulai dari bulan September 2016 hingga bulan September 2018
  * Penjualan tertinggi ada di bulan ke-13 yaitu November 2017
  * Penjualan terendah ada di bulan kedua yaitu Desember 2016

### Analisis RFM (Recency, Frequency, and Monetary)
Analisis RFM ini digunakan untuk menjawab pertanyaan 6 sampai 8

#### Persiapan Data

```python
# Menggabungkan data orders dengan data payments
orders_with_payments = pd.merge(df_orders, df_order_payments, on='order_id')
orders_with_payments
```

#### Hitung Metrik RFM

```python
# Mengelompokkan berdasarkan 'customer_id' dan mencari tanggal pembelian terakhir
last_purchase_per_customer = orders_with_payments.groupby('customer_id')['order_purchase_timestamp'].max().reset_index()

# Mencari tanggal pembelian terakhir dari seluruh pelanggan
latest_purchase_date = last_purchase_per_customer['order_purchase_timestamp'].max()

print(f"Tanggal terakhir pelanggan melakukan transaksi adalah: {latest_purchase_date}")
```

```python
# asumsikan today adalah hari terakhir pembelian dari seluruh pelanggan
today = pd.to_datetime(latest_purchase_date)
```

```python
# Menghitung Recency, Frequency, dan Monetary
rfm = orders_with_payments.groupby('customer_id').agg({
    'order_purchase_timestamp': lambda x: (today - x.max()).days,  # Recency
    'order_id': 'count',  # Frequency
    'payment_value': 'sum'  # Monetary
}).reset_index()

rfm.columns = ['customer_id', 'recency', 'frequency', 'monetary']
rfm.sort_values(by="recency", ascending=True)
```
